## Format as per SQuAD data

```python
squad_data = {"data": []}
for entry in new_data:
    squad_entry = dict()  # Initialize squad_entry as an empty dictionary
    squad_entry["title"] = entry["heading"]
    squad_entry["paragraphs"] = []

    context = entry["extracted_text"] + " " + entry['summarized_text']
    paragraphs = dict()  # Initialize paragraphs as an empty dictionary
    paragraphs["context"] = context
    paragraphs["qas"] = []

    for i, q in enumerate(entry["questions"]):
        qas_entry = dict()  # Initialize qas_entry as an empty dictionary
        qas_entry["question"] = q["question"]
        qas_entry["id"] = f"{entry['page']}_{i}"
        qas_entry["answers"] = [
            {"text": q["answer"]}
        ]
        paragraphs["qas"].append(qas_entry)

    squad_entry["paragraphs"].append(paragraphs)
    squad_data["data"].append(squad_entry)

print(squad_data)
```
